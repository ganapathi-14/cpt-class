```python
import numpy as np
import matplotlib.pyplot as plt
# Set a seed for the random number generator to make the results reproducible (same output every time)
np.random.seed(20)
x = np.random.rand(50) * 100        # Generate 50 random x-values scaled to 0–100
y = np.random.rand(50) * 100        # Generate 50 random y-values scaled to 0–100
colors = np.random.rand(50)         # Generate 50 random values for color mapping
sizes = np.random.randint(50, 500, 50) # Generate 50 random sizes for the scatter points
# Create a scatter plot with color mapping and variable sizes
scatter = plt.scatter(
    x, y,
    c=colors,             # Color values
    s=sizes,              # Size of each point
    cmap='hot',           # Colormap used for the colors
    edgecolors='black',   # Outline color of points
    alpha=0.6             # Transparency of points-makes the points slightly transparent (60% opacity)
)
# Add a color bar (legend for colors)
plt.colorbar(scatter, label="Color Scale")
# Add axis labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Random Scatter Plot")
# Add grid for better readability
plt.grid()
plt.show()
```

```python
#Sctter-3D
#Run this code in IDLE or VS-Code.Bcoz In this jupyter Not Rotating Like 3D animated
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
np.random.seed(20)
x = np.random.rand(50) * 100
y = np.random.rand(50) * 100
z = np.random.rand(50) * 100
colors = np.random.rand(50)
sizes = np.random.randint(50, 500, 50)
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
scatter = ax.scatter(
    x, y,z,
    c=colors,
    s=sizes,
    cmap='hot',
    edgecolors='black',
    alpha=0.7
)
fig.colorbar(scatter,ax=ax,label="Color Scale")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
plt.title("3-D Random Scatter Plot")
plt.grid()
plt.show()
```

```python
#Multi Scatter 3D
#Run this code in IDLE or VS-Code.Bcoz In this--jupyter Not Rotating Like 3D animated
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
np.random.seed(42)
x1 = np.random.rand(50) * 100
y1 = np.random.rand(50) * 100
z1 = np.random.rand(50) * 100
colors1 = np.random.rand(50)
sizes1 = np.random.randint(50, 500, 50)

x2 = np.random.rand(40) * 100
y2 = np.random.rand(40) * 100
z2= np.random.rand(40) * 100
colors2 = np.random.rand(40)
sizes2 = np.random.randint(50, 500, 40)

fig=plt.figure(figsize=(12,8))
ax=fig.add_subplot(111,projection='3d')

scatter1= ax.scatter(
    x1,y1,z1,
    c=colors1,
    s=sizes1,
    cmap='viridis',
    edgecolors='black',
    marker='o',
    alpha=0.7,
    label='Dataset 1'
)

scatter2= ax.scatter(
    x2,y2,z2,
    c=colors2,
    s=sizes2,
    cmap='plasma',
    edgecolors='black',
    marker='^',
    alpha=0.8,
    label='Dataset 2'
)
cbar1=fig.colorbar(scatter1,ax=ax,shrink=0.5,pad=0.1)
cbar1.set_label('Viridis')
cbar2=fig.colorbar(scatter2,ax=ax,shrink=0.5,pad=0.09)
cbar2.set_label('Plasma')
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
plt.title("3-D Random Scatter Plot with 2 Datsets")
plt.legend()
plt.grid()
plt.show()
```

```python
#Multiple 3D Histogram
#Run this code in IDLE or VS-Code.Bcoz In this--jupyter Not Rotating Like 3D animated
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
np.random.seed(42)
data1 = np.random.randint(1, 100, 50)
data2 = np.random.randint(1, 100, 50)
data3 = np.random.randint(1, 100, 50)

bins=10
count1,bin_edges=np.histogram(data1,bins=bins,range=(0,100))
count2,_=np.histogram(data2,bins=bins,range=(0,100))
count3,_=np.histogram(data3,bins=bins,range=(0,100))
bin_centers=0.5*(bin_edges[:-1]+bin_edges[1:])

fig=plt.figure(figsize=(12,8))
ax=fig.add_subplot(111,projection='3d')
dx=dy=(bin_edges[1]-bin_edges[0])

x1=bin_centers-dx
x2=bin_centers
x3=bin_centers+dx

y1=np.zeros_like(x1)
y2=np.zeros_like(x2) * 1.5
y3=np.zeros_like(x3) * 3

ax.bar3d(x1,y1, np.zeros_like(count1),dx,dy,count1,
         color='red',alpha=0.8,label='Dataset 1')
ax.bar3d(x2,y2, np.zeros_like(count2),dx,dy,count2,
         color='green',alpha=0.8,label='Dataset 2')
ax.bar3d(x3,y3, np.zeros_like(count3),dx,dy,count3,
         color='blue',alpha=0.8,label='Dataset 3')
ax.set_xlabel('Values')
ax.set_ylabel('Frequency')
plt.title('Multiple histograms overlapping')
ax.set_yticks([0,1.5,3])
ax.set_yticklabels(['Dataset 1', 'Dataset 2', 'Dataset 3'])

plt.legend()
plt.show()
```

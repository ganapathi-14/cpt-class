```python
import matplotlib.pyplot as plt

labels = ['Apples', 'Oranges', 'Kiwi', 'Watermelon']
prizes = [30, 25, 35, 10]
colors = ['red', 'orange', 'green', 'yellow']
explode = (0, 0, 0, 0.2)

plt.pie(
    prizes,
    labels=labels,
    colors=colors,
    startangle=90,
    explode=explode,
    shadow=True,
    autopct='%1.1f%%'
)

plt.title("Fruits Prizes")
plt.axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['C', 'C++', 'JAVA', 'PYTHON', 'DataScience']
prizes = [400, 450, 1100, 700, 1350]
colors = ['blue', 'grey', 'brown', 'magenta', 'green']
explode = (0, 0, 0.2, 0, 0.2)

def label_prizes(pct, all_vals):
    absolute = int(round(pct/100. * sum(all_vals)))
    return f"Rs.{absolute}"

plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    prizes,
    labels=labels,
    colors=colors,
    explode=explode,
    shadow=True,
    autopct=lambda pct: label_prizes(pct, prizes),
    pctdistance=0.6,
    startangle=90,
    textprops={'fontsize': 12, 'fontweight': 'bold', 'color': 'black'}
)

plt.title("Books Prizes", fontsize=16, fontweight='bold')
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.patches import Wedge
from matplotlib.collections import PatchCollection

labels = ['C', 'C++', 'JAVA', 'Python', 'DataScience']
prizes = [400, 450, 1100, 700, 1350]
colors = ['blue', 'grey', 'brown', 'magenta', 'green']

total = sum(prizes)
fracs = [p / total for p in prizes]  # fractions for pie slices
angles=[frac * 360 for frac in fracs]

fig=plt.figure(figsize=(8, 8))
ax=fig.add_subplot(111,projection='3d')
radius=1
height=0.3
start_angle=0
for i, angle in enumerate(angles):
    theta=np.linspace(np.deg2rad(start_angle),np.deg2rad(start_angle+angle),30)
    x=np.append([0],radius * np.cos(theta))
    y=np.append([0],radius * np.sin(theta))
    ax.plot_trisurf(x,y,height,color=colors[i],alpha=0.8)

    for j in range(len(theta)-1):
        xs=[radius *np.cos(theta[j]),radius*np.cos(theta[j+1])]
        ys=[radius*np.sin(theta[j]),radius*np.sin(theta[j+1])]
        zs=[0,0,height,height]
        verts=[list(zip(xs,ys,zs))]
ax.add_collection3d(plt.Poly3DCollection(verts,color=colors[i],alpha=0.8))
mid_angle=np.deg2rad(start_angle+angle/2)
ax.text(radius* 0.6 *np.cos(mid_angle),radius*0.6*np.sin(mid_angle),
        height + 0.05,f"{label[i]}:RS.{prizes[i]}",ha='center',va='bottom',
        fontsize=10,color='black')
start_angle +=angle 
ax.set_box_aspect([1,1,0.5])
ax.axis('off')
plt.title("3d pie")
plt.show()
    
```

```python
#Run this code in IDLE or VS-Code.Bcoz In this--jupyter Not Rotating Like 3D animated
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

labels = ['C', 'C++', 'JAVA', 'Python', 'DataScience']
prizes = [400, 450, 1100, 700, 1350]
colors = ['blue', 'grey', 'brown', 'magenta', 'green']

total = sum(prizes)
fracs = [p / total for p in prizes]
angles = [frac * 360 for frac in fracs]

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

radius = 1
height = 0.3
start_angle = 0

for i, angle in enumerate(angles):
    # Top surface of slice
    theta = np.linspace(np.deg2rad(start_angle), np.deg2rad(start_angle + angle), 30)
    x = np.append([0], radius * np.cos(theta))
    y = np.append([0], radius * np.sin(theta))
    z = np.ones_like(x) * height
    ax.plot_trisurf(x, y, z, color=colors[i], alpha=0.8)

    # Side wall
    verts = []
    for j in range(len(theta) - 1):
        xs = [radius * np.cos(theta[j]), radius * np.cos(theta[j+1]),
              radius * np.cos(theta[j+1]), radius * np.cos(theta[j])]
        ys = [radius * np.sin(theta[j]), radius * np.sin(theta[j+1]),
              radius * np.sin(theta[j+1]), radius * np.sin(theta[j])]
        zs = [0, 0, height, height]
        verts.append(list(zip(xs, ys, zs)))
    ax.add_collection3d(Poly3DCollection(verts, color=colors[i], alpha=0.8))

    # Label
    mid_angle = np.deg2rad(start_angle + angle / 2)
    ax.text(radius * 0.6 * np.cos(mid_angle),
            radius * 0.6 * np.sin(mid_angle),
            height + 0.05,
            f"{labels[i]}: Rs.{prizes[i]}",
            ha='center', va='bottom',
            fontsize=10, color='black')

    start_angle += angle

ax.set_box_aspect([1, 1, 0.5])
ax.axis('off')
plt.title("3D Pie Chart", fontsize=16)
plt.show()
```

```python
#Run this code in IDLE or VS-Code.Bcoz In this--jupyter Not Rotating Like 3D animated
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Provided data
labels = ['C', 'C++', 'JAVA', 'Python', 'DataScience']
prizes = [400, 450, 1100, 700, 1350]
colors = ['blue', 'grey', 'brown', 'magenta', 'green']

total = sum(prizes)
fracs = [p / total for p in prizes]
angles = [frac * 360 for frac in fracs]

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
radius = 1
height = 0.3
start_angle = 0

for i, angle in enumerate(angles):
    theta = np.linspace(np.deg2rad(start_angle), np.deg2rad(start_angle + angle), 30)
    x_top = np.append([0], radius * np.cos(theta))
    y_top = np.append([0], radius * np.sin(theta))
    z_top = np.full_like(x_top, height)
    ax.plot_trisurf(x_top, y_top, z_top, color=colors[i], alpha=0.8)

    verts = []
    for j in range(len(theta) - 1):
        x1, y1 = radius * np.cos(theta[j]), radius * np.sin(theta[j])
        x2, y2 = radius * np.cos(theta[j+1]), radius * np.sin(theta[j+1])
        verts.append([(x1, y1, 0), (x2, y2, 0), (x2, y2, height), (x1, y1, height)])

    ax.add_collection3d(Poly3DCollection(verts, color=colors[i], alpha=0.8))

    mid_angle = np.deg2rad(start_angle + angle / 2)
    ax.text(radius * 0.6 * np.cos(mid_angle),
            radius * 0.6 * np.sin(mid_angle),
            height + 0.05,
            f"{labels[i]}: Rs.{prizes[i]}",
            ha='center', va='bottom',
            fontsize=10, color='black')

    start_angle += angle

ax.set_box_aspect([1, 1, 0.5])
ax.axis('off')
plt.title("3D Pie Chart")
plt.show()
```

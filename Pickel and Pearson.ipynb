```python
#PickleClass

import pickle
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def hi(self):
        return f"Hello, my name is {self.name}"

# Create an instance of Student
s = Student('Vijay', 34)

# Serialize the object to a file
with open('student.pkl', 'wb') as f:
    pickle.dump(s, f)

# Deserialize the object from the file
with open('student.pkl', 'rb') as f:
    loadeddata = pickle.load(f)

# Access attributes and methods
print(loadeddata.name)      
print(loadeddata.hi())      
```

```python
#Co-relation Pearson

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Set seed for reproducibility
np.random.seed(42)

# Create theta values from 0 to 6*pi
theta = np.linspace(0, 6 * np.pi, 500)

# Create radius values from 0.5 to 50
r = np.linspace(0.5, 50, 500)

# Generate normal noise with mean 0 and std deviation 2
n = np.random.normal(0, 2, 500)

# Calculate x and y with noise
x = r * np.cos(theta) + n
y = r * np.sin(theta) + n

# Calculate Pearson correlation coefficient and p-value
r_value, p_value = pearsonr(x, y)
print(f"Pearson r: {r_value:.3f}, p-value: {p_value:.3f}")

# Plotting
plt.figure(figsize=(7, 7))
plt.scatter(x, y, c=theta, cmap='plasma')
plt.colorbar(label='Theta (radians)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Noisy Spiral Scatter Plot')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

np.random.seed(42)

theta = np.linspace(0, 6 * np.pi, 500)
r = np.linspace(0, 50, 500) + np.random.normal(0, 2, 500)
x = r * np.cos(theta)
y = r * np.sin(theta)

r_value, p_value = pearsonr(x, y)

plt.figure(figsize=(7, 7))
plt.scatter(x, y, c=theta, cmap='plasma', s=5, alpha=0.8)
plt.title(f"Pearson plot r = {r_value:.3f}", fontsize=15)
plt.axis("equal")  # makes the x and y axes have equal scale (like 'look' in your code)
plt.colorbar(label="Angle theta")
plt.show()
```

```python
#Same above code in 3D
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from mpl_toolkits.mplot3d import Axes3D 
np.random.seed(42)
theta=np.linspace(0,6*np.pi,800)
r=np.linspace(0,50,800) +np.random.normal(0,2,800)
z=np.linspace(-10,10,800)
x=r*np.cos(theta)
y=r*np.sin(theta)

r_value,p_value=pearsonr(x,y)
fig= plt.figure(figsize=(8,8))
ax= fig.add_subplot(111, projection='3d')
sc=ax.scatter(x,y,z,c=theta,cmap='plasma',s=5,alpha=0.8)
plt.title(f"Pearson plot r={r_value:.3f}",fontsize=15)
plt.colorbar(sc,ax=ax,label="Angle theta")
plt.show()
```
